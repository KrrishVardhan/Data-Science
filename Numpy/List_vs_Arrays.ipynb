```python
# Using Python Lists
import time

size = 1_000_000_0
l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [x+y for x,y in zip(l1,l2)]
end = time.time()
print(end - start)
```

```python
# Using Numpy Arrays
# WAAAY Faster than Python Lists
import numpy as np
import time

size = 1_000_000_0
a1 = np.array(list(range(size)))
a2 = np.array(list(range(size)))

start = time.time()
add = a1 + a2
end = time.time()
print(end - start)
```

```python
import sys
import numpy as np

list_data = list(range(1000))
numpy_data = np.array(list_data)

list_size = sys.getsizeof(list_data) + sum(sys.getsizeof(x) for x in list_data)
numpy_size = numpy_data.nbytes

print("Python list size:", list_size)
print("NumPy array size:", numpy_size)
```
