## BE VERY CAREFUL WITH <b style="color: red;">WHILE</b> LOOPS!

```python
a = 1
while a<12:
    print(a)
    a += 1
```

```python
l = [123,3,2,3,23,2,32,3,2,24,3,4,]
for i in l:
    print(i)
```

```python
fruits = ["apple","banana","orange","grape","strawberry","blueberry","pineapple","mango","kiwi","peach","pear","cherry"]
for i in fruits:
    if(i == "blueberry"):
        break
    print(i)
```

```python
fruits = ["apple","banana","orange","grape","strawberry","blueberry","pineapple","mango","kiwi","peach","pear","cherry"]
for i in fruits:
    if(i == "blueberry"):
        continue # This will skip "blueberry" from the iteration and continue forward
    print(i)
```
