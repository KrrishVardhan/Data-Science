```python
class Employee:
    company = "Microsoft"
    def __init__(self, name, salary):  # this is a constructor
        self.name = name
        self.salary = salary
    def PrintDetails(self):
        print(f"The company name of {self.name} is {self.company} with salary ${self.salary}")

    @staticmethod # Decorator
    def PrintTime():
        print(f"The time is Now") # standalone method that takes no self or cls as parameter
k = Employee("Krrish", 45000)
k.PrintDetails()
k.PrintTime()
```

```python
class Employee:
    company = "Microsoft"
    def __init__(self, name, salary):  # this is a constructor
        self.name = name
        self.salary = salary
    def PrintDetails(self):
        print(f"The company name of {self.name} is {self.company} with salary ${self.salary}")

    @staticmethod # Decorator
    def PrintTime():
        print(f"The time is Now") # standalone method that takes no self or cls as parameter

    @classmethod
    def PrintCompany(cls):
        print(f"The company is {cls.company}")
        # print(f"The name is {cls.name}") This will give error
k = Employee("Krrish", 45000)
k.PrintDetails()
k.PrintTime()
k.PrintCompany()
```
