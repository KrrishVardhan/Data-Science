```python
def http_status(req):
    match req:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown Status"
```

```python
code = 200
```

```python
http_status(code)
```

```python
newcode = 500
http_status(newcode)
```
