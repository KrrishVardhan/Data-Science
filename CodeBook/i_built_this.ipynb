```python
import json
# Load data

def load_data(filename):
    with open(filename, "r") as file:
        data = json.load(file)
    return data
# Pages you might like 
def pages_you_like(user_id, data):
    for user in data["users"]:
        if user["id"] != user_id:
            continue
        return user["liked_pages"]
def pages_you_might_like(pages_you_like, data):
    suggestions = []
    for user in data["users"]:
        if set(pages_you_like).intersection(set(user["liked_pages"])):
            suggestions.extend(set(user["liked_pages"]))
    final_suggestions = [page for page in set(suggestions) if page not in pages_you_like]
    return final_suggestions
```

```python
data = load_data("massive_data.json")
user_id = 1
liked_pages = pages_you_like(user_id,data)
pages = pages_you_might_like(liked_pages, data)
print(f"for User {user_id} ->You might like these pages as well: {pages}")
```
